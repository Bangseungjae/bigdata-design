```python
breast_cancer_df.replace({'diagnosis': {'M': 0}}, inplace=True)
breast_cancer_df.replace({'diagnosis': {'B': 1}}, inplace=True)
breast_cancer_df.head(5)
```

```python
a = ['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean']
b = ['diagnosis', 'symmetry_mean', 'fractal_dimension_mean', 'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se', 'compactness_se', 'concavity_se']
c = ['diagnosis', 'concave points_se', 'symmetry_se', 'fractal_dimension_se', 'radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst', 'smoothness_worst', 'concavity_worst', 'concave points_worst', 'symmetry_worst', 'fractal_dimension_worst']

sns.heatmap(breast_cancer_df[c].corr(), annot=True)

# 'radius_mean', 'perimeter_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean',
# 'radius_se', 'perimeter_se', 'area_se', 'radius_worst', 'perimeter_worst', 'area_worst', 'concavity_worst', 'concave points_worst'
```

```python
# x_columns = ['radius_mean', 'perimeter_mean', 'smoothness_mean', 'concavity_mean', 'symmetry_mean', 'radius_se', 'perimeter_se',
#              'smoothness_se', 'concavity_se', 'symmetry_se', 'radius_worst', 'perimeter_worst', 'smoothness_worst', 'concavity_worst', 'symmetry_worst']

x_columns = ['radius_mean', 'perimeter_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean',
             'radius_se', 'perimeter_se', 'area_se', 'radius_worst', 'perimeter_worst', 'area_worst', 'concavity_worst', 'concave points_worst']
y_columns = 'diagnosis'

X = breast_cancer_df[x_columns]
Y = breast_cancer_df[y_columns]

import numpy as np
X = np.array(X)
Y = np.array(Y)
print(X.shape)
print(Y.shape)
```

```python
from sklearn.metrics import *
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, shuffle=False, random_state=1222)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
clf = DecisionTreeClassifier(max_depth=4)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred)

print("Acc: {}%".format(np.round(acc * 100, 3)))
```
